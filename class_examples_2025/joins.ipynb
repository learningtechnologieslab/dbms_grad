```python
import pandas as pd
```

```python
orders = pd.read_csv('orders.csv')
customers = pd.read_csv('customers.csv')
payments = pd.read_csv('payments.csv')
```

```python
orders.head()
```

```python
customers.head()
```

### Inner Join

```python
results = pd.merge(customers, orders, how='inner', left_on='customerNumber', right_on='customerNumber')
results.head()
```

### Left Join

```python
results = pd.merge(customers, orders, how='left', left_on='customerNumber', right_on='customerNumber')
results[['customerNumber', 'orderNumber']].isnull().sum()
```

### Right Join

```python
results = pd.merge(orders, customers, how='right', left_on='customerNumber', right_on='customerNumber')
results[['customerNumber', 'orderNumber']].isnull().sum()
```

### FullL Join

```python
results = pd.merge(orders, customers, how='outer', left_on='customerNumber', right_on='customerNumber')
results[['customerNumber', 'orderNumber']].isnull().sum()
```
