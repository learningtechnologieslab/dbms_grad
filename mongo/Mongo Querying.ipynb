```python
from pymongo import MongoClient
from bson.json_util import dumps
import json
```

```python
client = MongoClient('localhost', 27017)
```

```python
database = client['videoreviews']
```

```python
collection = database.get_collection('video_movies')
```

```python
data = database.collection.find({'year' : 2014})
```

### Equality

```python
for doc in collection.find({"year" : 2014, "type": "movie", "imdb" : "tt2305051"}):
    print(doc)
```

### Less Than 

_{"key" :{$lt:"value"}}_


```python
for doc in collection.find({"year":{"$lt":2014}}):
    print(doc)
```

### Less Than or Equal To

_{"key" :{$lte:"value"}}_

```python
for doc in collection.find({"year":{"$lte":2014}}):
    print(doc)
```

### Greater Than

_{"key" :{$gt:"value"}}_

```python
for doc in collection.find({"year":{"$gt":2014}}):
    print(doc)
```

### Greater Than or Equal To
_{"key" :{$gte:"value"}}_

```python
for doc in collection.find({"year":{"$gte":2014}}):
    print(doc)
```

### Not Equals
_{"key":{$ne: "value"}}_

```python
for doc in collection.find({"year":{"$ne":2014}}):
    print(doc)
```

### in

```python
for doc in collection.find({"year":{"$in":[1999, 2000, 2001]}}):
    print(doc)
```

### OR

```python
for doc in collection.find({"$or":[{"year":2014},{"title": "Big Daddy"}]}):
    print(doc)
```

```python
for doc in collection.find({"$or":[{"year":2014},{"title": "Big Daddy"}]}).sort("year").limit(5):
    print(doc)
```
