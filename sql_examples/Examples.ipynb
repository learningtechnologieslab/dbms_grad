```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df['frame'].unique()
```

```python
df.groupby('frame').agg({'weight':'mean'}).query('weight > 200')
```

```python
df.head()
```

```python
data1 = df[['id','chol','age']].loc[0:4]
data1
```

```python
data2 = df[['id','gender','weight','height']].loc[1:5]
data2
```

```python
result = pd.merge(data1, data2, how='inner', left_on='id', right_on='id')
result
```

```python
result = pd.merge(data1, data2, how='left', left_on='id', right_on='id')
result
```

```python
result = pd.merge(data1, data2, how='right', left_on='id', right_on='id')
result
```

```python
result = pd.merge(data1, data2, how='outer', left_on='id', right_on='id')
result
```
